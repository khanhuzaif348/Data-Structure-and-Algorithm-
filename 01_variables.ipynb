```python
# lets Start with Variables in Python
# Variables are used to store data in Python
# A variable is created when you assign a value to it


```

```python
# Example of variable declaration and assigning values

x = 5           # Integer variable
y = 3.14        # Float variable    
name = "Alice"  # String variable
is_active = True  # Boolean variable
# You can print the value of a variable using the print() function
print(x)
print(y)    
print(name)
print(is_active)
```

```python
#Naming Conventions:
# Variable names must start with a letter or an underscore (_)
# Variable names can contain letters, numbers, and underscores
# Variable names are case-sensitive (e.g., myVar and myvar are different variables) 
# Avoid using Python reserved keywords as variable names (e.g., if, else, while, etc.)
# Use descriptive names for variables to make your code more readable
# Example of good variable names
user_age = 25
total_price = 99.99
is_logged_in = False
# Example of bad variable names
a = 10
b = 20.5
c = "Hello"
# d = True
# --- IGNORE ---
# # This is to prevent execution errors in the notebook environment

pass
#INvalid code below
# x y = 10  # Invalid: space in variable name
# 2nd_value = 20  # Invalid: variable name cannot start with a number
```

```python
#UNderstand variable types
print(type(x))          # Output: <class 'int'>
print(type(y))          # Output: <class 'float'>   
print(type(name))       # Output: <class 'str'>
print(type(is_active))  # Output: <class 'bool'>
# You can change the value of a variable by reassigning it
x = 10
print(x)  # Output: 10
y = 6.28
print(y)  # Output: 6.28
name = "Bob"
print(name)  # Output: Bob
is_active = False
print(is_active)  # Output: False
# You can also assign the value of one variable to another variable
a = x
b = y
c = name
d = is_active


        
```

```python
# Type  checking and Conversion
print(type(a))  # Output: <class 'int'>
print(type(b))  # Output: <class 'float'>
print(type(c))  # Output: <class 'str'>
print(type(d))  # Output: <class 'bool'>
```

```python
age = 30
age_str = str(age)  # Convert integer to string
print(age_str)      # Output: "30"
print(type(age_str))  # Output: <class 'str'>
```

```python
#type of conversion
age_float = float(age)  # Convert integer to float
print(age_float)        # Output: 30.0
print(type(age_float))   # Output: <class 'float'>
```
