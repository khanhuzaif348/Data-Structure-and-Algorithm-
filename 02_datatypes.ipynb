# Introduction of Data Types in Python


```python
# Introduction of Data Types in Python
# Importance of Datatypes in programming
# Type checking and conversion
# Dynamic typing in Python


# imporatantance of Data Types
# Data types are essential in programming as they define the type of data a variable can hold
# They help in determining the operations that can be performed on the data
# Python is a dynamically typed language, meaning you don't need to declare the data type of a variable explicitly
# The interpreter infers the type based on the assigned value
# Common Data Types in Python  
# Integer (int): Whole numbers, e.g., 1, -5, 100
# Float (float): Decimal numbers, e.g., 3.14, -0.001
# String (str): Text, e.g., "Hello", 'Python'
# Boolean (bool): True or False values
```

```python
# Advanced Data type
# List: Ordered, mutable collection of items, e.g., [1, 2, 3], ['a', 'b', 'c']
my_list = [1, 2, 3, 4, 5]
print(type(my_list))  # Output: <class 'list'>
# Tuple: Ordered, immutable collection of items, e.g., (1, 2, 3), ('a', 'b', 'c')
my_tuple = (1, 2, 3, 4, 5)
print(type(my_tuple))  # Output: <class 'tuple'>
# Dictionary: Unordered collection of key-value pairs, e.g., {'name': 'Alice
my_dict = {'name': 'Alice', 'age': 25}
print(type(my_dict))  # Output: <class 'dict'>
# Set: Unordered collection of unique items, e.g., {1, 2,
my_set = {1, 2, 3, 4, 5}
print(type(my_set))  # Output: <class 'set'>
# Type Checking and Conversion
# You can check the type of a variable using the type() function
x = 10
print(type(x))          # Output: <class 'int'>
y = 3.14
print(type(y))          # Output: <class 'float'>
name = "Alice"
print(type(name))       # Output: <class 'str'>
is_active = True
print(type(is_active))  # Output: <class 'bool'>
# You can convert between data types using built-in functions
# Converting int to float
x_float = float(x)
print(x_float)          # Output: 10.0
print(type(x_float))    # Output: <class 'float'>
# Converting float to int
y_int = int(y)
print(y_int)            # Output: 3
print(type(y_int))      # Output: <class 'int'>
# Converting int to string
x_str = str(x)
print(x_str)           # Output: "10"
print(type(x_str))     # Output: <class 'str'>
# Dynamic Typing in Python
# In Python, you can reassign variables to different data types
var = 10               # Initially an integer
print(type(var))      # Output: <class 'int'>
var = "Hello"         # Now a string
print(type(var))      # Output: <class 'str'>
var = 3.14           # Now a float
print(type(var))      # Output: <class 'float'>
var = True            # Now a boolean
print(type(var))      # Output: <class 'bool'>
# This flexibility allows for rapid development and prototyping





```

```python
# Practice Example
# Create variables of different data types and print their types
a = 42                  # Integer
b = 2.718               # Float
c = "Python Programming" # String
d = False               # Boolean   
print(type(a))  # Output: <class 'int'>
print(type(b))  # Output: <class 'float'>
print(type(c))  # Output: <class 'str'>
print(type(d))  # Output: <class 'bool'>
```
