#### Eg- 1 : Temperature Conversion

```python
#(32°F − 32) × 5/9 f-c
#0°C × 9/5) + 32
def temp_cenversion(temp,unit):
    ''' This fucntion are used to convert a temp '''
    if unit=='f':
        return (temp-32) * 5/9
    elif unit=='c':
        return temp * 9/5 + 32

```

```python
print(temp_cenversion(unit='f',temp = 77))
print(temp_cenversion(unit='c',temp = 25))
```

## Password Strength Checker

```python
#  ex -  Password strength checker 

def is_password_srrong(password):
    """ this fucntion check  the password strength"""
    # len must be greater or equal to 8 
    if len(password) < 8:
        return False 
    if not any(char.isdigit() for char in  password):
        return False
    if not any(char.islower() for char in password):
        return False 
    if not any(char.isupper() for char in password):
        return  False
    if not any(char in '!@#$%^&*()_+' for char in password):
        return False
    return True    
        
```

```python
is_password_srrong('MOHDhuz)@786')
```

```python
#3 Calculate the total Cost of Items in shopping Cart 
def calculate_total_cost(cart):
    total_cost=0
    for item in cart:
        total_cost = total_cost + item['price'] * item['quantity'] 

    return round(total_cost,2)

cart=[
    {'price':0.5,'quantity': 4} , 
    {'price':0.9,'quantity': 6} ,
    {'price':0.2,'quantity': 4}
]
    
        
```

```python
calculate_total_cost(cart)
```

```python
# Example 4 : Check If a String is palindrome 
def is_palindrome(s):
    s= s.lower().replace(" ","")
    return s==s[::-1]
print(is_palindrome(" ABBA"))
    
    
```

```python
## Ex -  5  Calculate the factorials of a number using recursion 

def factorial(n):
    if n==0:
        return 1
    else:
        return n* factorial(n-1)
```

```python
factorial(12)
```
