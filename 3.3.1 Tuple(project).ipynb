## Example 1. MAnage A to DO List

```python
to_do_list = ["Buy grocesries","buy Vegetable" ,"pay house Rent","pay bill"]


## Adding items to the list
to_do_list.append("Schedule  meeting")
to_do_list.append("go for  a run")

# Remove for task completed 

to_do_list.remove("Buy grocesries")

#checking if a task in to to_do_list
if "pay house Rent" in to_do_list:
    print("Hey ! don't forget the payment of rent ")

#print remaining task 
print("remaining task")
for task in to_do_list:
    print(task)
    
```

## Example 2 : Organizing Student grades

* Create a list to store and calculate average grades for students

```python
# list of grades 
grades =[56,45,89,87]
# adding new grade
grades.append(95)
print("new grades added")

#calculating a average of grade

average_grade = sum(grades) / len(grades)
print(f' average grades :  {average_grade}')


# finding the max , min grades  
high_grade,min_grade  =  max(grades ) , min( grades) 
print(f' highest grades  : {high_grade}')
print(f'minimum grades   : {min_grade}')
```

## Customer Feedback 

```python
#collecting user feedback 
feedback  = ["Great Service","very satisfied" , "could be better" , "excellenet experience "]

# adding a new  feeedback 
feedback.append(" Good choice")
print("New feedback Added")
print("------------------")

# counting Feedback 
count_fb =  sum([ 1 for fb in feedback if "great" in fb.lower() or "good" in fb.lower() ])
print(f'total no of feedback {count_fb}')
print("------------------")
# Print feedback 


print(" remaining Feedback  ")
print("------------------")
for i in feedback:
    print(f' - {i}')
```
