```python
## creating a tuples 
empty_tuple = ()
print(empty_tuple)
print(type(empty_tuple))
```

```python
# creating list using tuple
lst  = list()
print(type(lst))
tpl = tuple()
print(type(tpl))
```

```python
## add some elemnt
numbers = tuple([1,2,3,4,5])
numbers[0] # indexing 
numbers[0:2] #slicing 

```

```python
# Tuple Operations

# Concatenation
tuple1 = (1, "Hello", 3)
tuple2 = (4, 5, 6)
combined_tuple = tuple1 + tuple2
print(combined_tuple)  # Output: (1, 2, 3, 4, 5, 6)
```

```python
# Repitation 
tuple1*3 # 3 times 
```

```python
## Immutability of Tuples

# Tuples are immutable, meaning their elements cannot be changed after creation.
# The following line would raise an error:  
# tuple[0] = 10  # This will raise a TypeError 
tuple[0] = 10

```

```python
# packing  -->  

pack_tuple = 1,2,3,"hello world"
print(pack_tuple)
```

```python
a ,b ,c , d = pack_tuple
print(a)
print(b)
print(c)
print(d)
```

```python
# Unpacking packing 
numbers =  (1,2,3,4,5)
*first,second,third =  numbers
print(first)
print(third)
```

```python
# Nested Tuples
nested_tuple = (1, (2, 3), (4, (5, 6)))
print(nested_tuple[1])        # Output: (2, 3)
print(nested_tuple[2][1])     # Output: (5, 6)
print(nested_tuple[2][1][0])  # Output: 5
```

```python
# Tuple Comprehensions
squared_tuple = tuple(x**2 for x in range(5))
print(squared_tuple)  # Output: (0, 1, 4, 9, 16)

```

```python
# Example 1: Using Tuples as Dictionary Keys
location_coords = {
    (40.7128, -74.0060): "New York",
    (34.0522, -118.2437): "Los Angeles",
    (41.8781, -87.6298): "Chicago"
}
print(location_coords[(40.7128, -74.0060)])  # Output: New York
```
